# Context

* Consider there is a submarine and a war going on between two countries. Submarine of a country is going to another country. The enemy country has placed some mines under water. There can also be rocks in the ocean. Our job is to make a model to identify if there's a rock or a mine underneath the submarine.

* Submarine sends a sound signal which is known as Sonar and detects the object details.

* In real cases, the data has been collected by sending and receiving the sonar signals to and from the metal and a rock. In laboratory setup, the metal cylinder represents the mine and rock is the ground.

# Methodology

* Collection of Data
* Pre-process the Data
* Train-test split
* Feed it to a ML model
  * Used Logistic Regression model
* Train our model
* Test our model

## Importing the dependencies

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

## Uploading the dataset

## Separating the Data and Labels

````python
X = sonar_data.drop(columns = 60, axis = 1)
y = sonar_data[60]
````

````python
X.head()
````

````python
y.head()
````

## Splitting the data into Training and Test data

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify = y, random_state = 42)
````

* ```test_size = 0.1``` means 10% of the dataset is test data and rest 90% dataset is training data.
* ```stratify``` ensure the equal splitting of R and M for test and train. It prevents skewing.
* ```random_state``` is used for specifying the splitting pattern.

````python
print(X.shape, X_train.shape, X_test.shape)
````

## Training the Model

### Logistic Regression Model
* This model works the best in the binary classification problem. The problem is to predict if the detected surface is a rock or a mine.
* This is a Supervised Learning model.

````python
model = LogisticRegression()
````

````python
model.fit(X_train, y_train)
````

## Model Evaluation

````python
# predicting the training data
X_train_prediction = model.predict(X_train)
````

````python
# calculating the accuracy_score
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
print(f'Accuracy on training data: {training_data_accuracy}')
````

````python
# predicting the testdata
X_test_prediction = model.predict(X_test)
````

````python
# calculating the accuracy_score
testing_data_accuracy = accuracy_score(X_test_prediction, y_test)
print(f'Accuracy on testing data: {testing_data_accuracy}')
````

## Random Forest Algorithm
* This model can capture non-linear relationships which the logistic regression cannot capture.

````python
# n_estimators is the number of trees in the forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
````

````python
# Fitting the model
rf_model.fit(X_train, y_train)
````

### Evaluating the model

````python
# predicting the training data
X_train_prediction_rf = rf_model.predict(X_train)
````

````python
# calculating the accuracy_score for training data
training_data_accuracy_rf = accuracy_score(X_train_prediction_rf, y_train)
print(f'Accuracy on training data: {training_data_accuracy_rf}')
````

````python
# predicting the testing data
X_test_prediction_rf = rf_model.predict(X_test)
````

````python
# making predictions
y_rf_pred = rf_model.predict(X_test)
````

# Advanced Prediction

## Applying SMOTE (Synthetic Minority Over-sampling Technique)

````python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Apply SMOTE to the Training data ONLY
# k_neighbors=5 is default; you can lower it if a class is extremely small
smote = SMOTE(random_state=42, k_neighbors=5)
X_train_reshaped, y_train_reshaped = smote.fit_resample(X_train, y_train)

print(f"Resampled training shape: {Counter(y_train_reshaped)}")
````

````python
# Now train your model on the reshaped data

rf_model_2 = RandomForestClassifier()
rf_model_2.fit(X_train_reshaped, y_train_reshaped)
````

````python
# making predictions
y_rf_pred_2 = rf_model_2.predict(X_test)
````

````python
print(f'Accuracy of Random Forest Classifier after applying SMOTE is: {accuracy_score(y_test, y_rf_pred_2)}')
````

## GridSearchCV

````python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from imblearn.pipeline import Pipeline
````

````python
# Define the pipeline: First SMOTE, then Classifier
model_pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(random_state=42))
])
````

````python
from scipy.stats import randint

param_dist = {
    'classifier__n_estimators': randint(100, 500),
    'classifier__max_depth': [None, 3, 5, 7, 10],
    'classifier__min_samples_split': randint(2, 20),
    'classifier__min_samples_leaf': randint(1, 10),
    'classifier__max_features': ['sqrt', 'log2', None]
}

# n_iter=50 means it will try 50 random combinations
random_search = RandomizedSearchCV(model_pipeline, param_distributions=param_dist,
                                   n_iter=50, cv=5, scoring='accuracy',
                                   random_state=42, n_jobs=-1)
random_search.fit(X_train, y_train)

print(f"Best RandomSearch Params: {random_search.best_params_}")
print(f"Best RandomSearch Score: {random_search.best_score_}")
````

````python
# Note the 'classifier__' prefix to tell the pipeline which step the params belong to
param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None, 5, 10],
    'classifier__min_samples_split': [2, 5],
    'classifier__max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(model_pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"Best GridSearch Params: {grid_search.best_params_}")
print(f"Best GridSearch Score: {grid_search.best_score_}")
````

````python
# Extract the best model
# final_model = random_search.best_estimator_

# If GridSearchCV, use:
final_model = grid_search.best_estimator_

final_model.fit(X_train, y_train)
````

````python
from sklearn.metrics import classification_report, confusion_matrix

# Make predictions
final_preds = final_model.predict(X_test)

# Check the results
print(f"Final Optimized Accuracy: {accuracy_score(y_test, final_preds)}")
print("\n--- Detailed Report ---")
print(classification_report(y_test, final_preds))
````

````python
# See exactly which ones it got wrong (Rock vs Mine)
print("\n--- Confusion Matrix ---")
print(confusion_matrix(y_test, final_preds))
````

````python
# reshape the numpy array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, 60)

prediction = final_model.predict(input_data_reshaped) # output is a list
print(prediction)

if (prediction[0] == 'R'):
  print('The object is a Rock')
else:
  print('The object is a Mine')
````
